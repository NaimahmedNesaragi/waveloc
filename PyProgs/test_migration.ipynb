After trying the above calculation with many different variants, the results always stays the same.  Results for y and z are similar.  What is going on here?  Why is the time-dimension so fundamentally different from the two space dimensions?  Can we really get away with interpreting the stack volume as a probability density distribution?

**Hypothesis1** : Let's consider a small time-slice around exp_t, and do the 4D integration around that.

```python
# Get the indexes for the t-arrayit0=np.int(np.round(exp_t/dt))ist0=np.int(np.round(np.sqrt(var_t)/dt))# Take 1-sigma time sliceit_left=it0-ist0it_right=it0+ist0t_slice=t[it_left:it_right]# extract stack and normalizestack_slice=stack_grid[:,:,:,it_left:it_right]stack_slice_int=compute_integral4D(stack_slice,x,y,z,t_slice)stack_slice_norm=stack_slice/stack_slice_int
```

```python
# Repeat the t calculation# Integrate over all space dimensions to get marginal probability density over tprob_t = si.trapz(si.trapz(si.trapz(stack_slice_norm,x=x,axis=0),x=y,axis=0),x=z,axis=0)exp_t = si.trapz(t_slice*prob_t,x=t_slice,axis=0)var_t = si.trapz((t_slice-exp_t)*(t_slice-exp_t)*prob_t,x=t_slice,axis=0)print '<t0> = %.3f pm %.3f s'%(exp_t,np.sqrt(var_t))plt.plot(t_slice,prob_t)plt.title('Marginal probability over t')
```
