```python
# Read in a stack file and set up its geometry correctly (hi res)nx,ny,nz,nt = 32,24,12,1338dx=dy=dz=0.25dt=0.01ix_true=16iy_true=8iz_true=6it_true=600stack_grid=np.fromfile('test_grid4D_hires.dat').reshape(nx,ny,nz,nt)x=np.arange(nx)*dxy=np.arange(ny)*dyz=np.arange(nz)*dzt=np.arange(nt)*dt
```

```python
# Normalize the stack for probability density calculationsstack_grid_int=compute_integral4D(stack_grid,x,y,z,t)print 'Integral over stack volume = %.3f'%stack_grid_intstack_grid_norm=stack_grid / stack_grid_int
```

```python
# Integrate over all space dimensions to get marginal probability density over tprob_t = si.trapz(si.trapz(si.trapz(stack_grid_norm,x=x,axis=0),x=y,axis=0),x=z,axis=0)exp_t = si.trapz(t*prob_t,x=t,axis=0)var_t = si.trapz((t-exp_t)*(t-exp_t)*prob_t,x=t,axis=0)print '<t0> = %.3f pm %.3f s'%(exp_t,np.sqrt(var_t))plt.plot(t,prob_t)plt.title('Marginal probability over t')
```

```python
# Try to do the same for xprob_x = si.trapz(si.trapz(si.trapz(stack_grid_norm,x=y,axis=1),x=z,axis=1),x=t,axis=1)exp_x = si.trapz(x*prob_x,x=x,axis=0)var_x = si.trapz((x-exp_x)*(x-exp_x)*prob_x,x=x,axis=0)print '<x0> = %.3f pm %.3f s'%(exp_x, np.sqrt(var_x))plt.plot(x,prob_x)plt.title('Marginal probability over x')
```

After trying the above calculation with many different variants, the results always stays the same.  Results for y and z are similar.  What is going on here?  Why is the time-dimension so fundamentally different from the two space dimensions?  Can we really get away with interpreting the stack volume as a probability density distribution?

**Hypothesis1** : Let's consider a small time-slice around exp_t, and do the 4D integration around that.

```python
# Get the indexes for the t-arrayit0=np.int(np.round(exp_t/dt))ist0=np.int(np.round(np.sqrt(var_t)/dt))# Take 1-sigma time sliceit_left=it0-1*ist0it_right=it0+1*ist0t_slice=t[it_left:it_right]# extract stack and normalizestack_slice=stack_grid[:,:,:,it_left:it_right]stack_slice_int=compute_integral4D(stack_slice,x,y,z,t_slice)stack_slice_norm=stack_slice/stack_slice_int
```

```python
# Repeat the t calculation# Integrate over all space dimensions to get marginal probability density over tprob_t = si.trapz(si.trapz(si.trapz(stack_slice_norm,x=x,axis=0),x=y,axis=0),x=z,axis=0)exp_t = si.trapz(t_slice*prob_t,x=t_slice,axis=0)var_t = si.trapz((t_slice-exp_t)*(t_slice-exp_t)*prob_t,x=t_slice,axis=0)print '<t0> = %.3f pm %.3f s - True t0 = %.3f'%(exp_t,np.sqrt(var_t),t[it_true])plt.plot(t_slice,prob_t)plt.vlines(exp_t,min(prob_t),max(prob_t),'b')plt.vlines(t[it_true],min(prob_t),max(prob_t),'r')plt.hlines(min(prob_t)+(max(prob_t)-min(prob_t))/2.0,exp_t-np.sqrt(var_t),exp_t+np.sqrt(var_t),'b')plt.title('Marginal probability over t')
```

```python
# and the x calculationprob_x = si.trapz(si.trapz(si.trapz(stack_slice_norm,x=y,axis=1),x=z,axis=1),x=t_slice,axis=1)exp_x = si.trapz(x*prob_x,x=x,axis=0)var_x = si.trapz((x-exp_x)*(x-exp_x)*prob_x,x=x,axis=0)print '<x0> = %.3f pm %.3f km - True x0 = %.3f'%(exp_x, np.sqrt(var_x),x[ix_true])plt.plot(x,prob_x)plt.vlines(exp_x,min(prob_x),max(prob_x),'b')plt.vlines(x[ix_true],min(prob_x),max(prob_x),'r')plt.hlines(min(prob_x)+(max(prob_x)-min(prob_x))/2.0,exp_x-np.sqrt(var_x),exp_x+np.sqrt(var_x),'b')plt.title('Marginal probability over x')
```

```python
# and the y calculationprob_y = si.trapz(si.trapz(si.trapz(stack_slice_norm,x=x,axis=0),x=z,axis=1),x=t_slice,axis=1)exp_y = si.trapz(y*prob_y,x=y,axis=0)var_y = si.trapz((y-exp_y)*(y-exp_y)*prob_y,x=y,axis=0)print '<y0> = %.3f pm %.3f km - True y0 = %.3f'%(exp_y, np.sqrt(var_y), y[iy_true])plt.plot(y,prob_y)plt.vlines(exp_y,min(prob_y),max(prob_y),'b')plt.vlines(y[iy_true],min(prob_y),max(prob_y),'r')plt.hlines(min(prob_y)+(max(prob_y)-min(prob_y))/2.0,exp_y-np.sqrt(var_y),exp_y+np.sqrt(var_y),'b')plt.title('Marginal probability over y')
```

```python
# and the z calculationprob_z = si.trapz(si.trapz(si.trapz(stack_slice_norm,x=x,axis=0),x=y,axis=0),x=t_slice,axis=1)exp_z = si.trapz(z*prob_z,x=z,axis=0)var_z = si.trapz((z-exp_z)*(z-exp_z)*prob_z,x=z,axis=0)print '<z0> = %.3f pm %.3f km - True z0 = %.3f'%(exp_z, np.sqrt(var_z), z[iz_true])plt.plot(z,prob_z)plt.vlines(exp_z,min(prob_z),max(prob_z),'b')plt.vlines(z[iz_true],min(prob_z),max(prob_z),'r')plt.hlines(min(prob_z)+(max(prob_z)-min(prob_z))/2.0,exp_z-np.sqrt(var_z),exp_z+np.sqrt(var_z),'b')plt.title('Marginal probability over z')
```
